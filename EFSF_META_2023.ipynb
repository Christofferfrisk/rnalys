```python
import os
from pathlib import Path

def search_files(directory, search_terms):
    matches = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            file_path = Path(root) / file
            try:
                with open(file_path, 'r', encoding='utf-8', errors='ignore') as f:
                    contents = f.read()
                    # Check if both search terms are in file
                    if all(term in contents for term in search_terms):
                        matches.append(file_path)
            except Exception as e:
                print(f"Error reading {file_path}: {e}")
    return matches

# Define the directory to search and the terms
search_directory = '~/Dropbox/'
search_terms = ['SLL59', 'Gender']

# Perform the search
found_files = search_files(search_directory, search_terms)
```
